**Homework 7.1**

Write code to search the Brown Corpus for particular words and phrases according to tags,
to answer the following questions:




(a) Produce an alphabetically sorted list of the distinct words tagged as MD.


```python
# add code
import nltk 
from nltk.corpus import brown 

brown_words = brown.words()
patterns = [(r'.*ould$', 'MD')]
regexp_tagger = nltk.RegexpTagger(patterns)
tags = regexp_tagger.tag(brown_words)
word_lst1 = list(set([a.lower() for (a,b) in tags if b =="MD"]))
```

```python
sorted_alphabetically = sorted(word_lst1, key=lambda tup: tup.lower())
print(sorted_alphabetically)
```

(b) Identify words that can be plural nouns or third person singular verbs (e.g. deals, flies).


```python
patterns2 = [(r'.*s$', 'NNS'), (r'.*es$', 'VBZ')]
regexp_tagger2 = nltk.RegexpTagger(patterns2)
tags2 = regexp_tagger2.tag(brown_words)
words_lst2 = list(set([a.lower() for (a,b) in tags2 if b =="NNS" or b =="VBZ"]))
sorted_alphabetically2 = sorted(words_lst2, key=lambda tup: tup.lower())
print(sorted_alphabetically2)
```
